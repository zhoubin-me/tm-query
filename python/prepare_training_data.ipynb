```python
entries_to_save = []
idx = 0
for entry in tqdm(data):
    image_name = os.path.join(entry['imageName'])
    try:
        image = Image.open(os.path.join('dset/imgs', image_name))
    except Exception as e:
        print(f"Error opening image {image_name}: {e}")
        continue

    width, height = image.size
    chinese_text = entry['chineseCharacter']
    words_in_mark = entry['wordsInMark']
    description = entry['descrOfDevice']

    if chinese_text is None:
        chinese_text = ""
    if words_in_mark is None:
        words_in_mark = ""
    if description is None:
        description = ""

    chinese_text = chinese_text.replace('\r', ' ').replace('\n', ' ').strip()
    words_in_mark = words_in_mark.replace('\r', ' ').replace('\n', ' ').strip()
    description = description.replace('\r', ' ').replace('\n', ' ').strip()

    if chinese_text is None and words_in_mark is None and description is None:
        continue
    conversations = [
        {"from": "human", "value": "<image>\nExtract Chinese characters, English words on this trademark image, and describe the trademark in short words."},
        {"from": "gpt", "value": "Chinese Characters: {}\nWords in mark: {}\nDescription: {}".format(chinese_text, words_in_mark, description)},
    ]

    entry_to_save = {
        "id": idx,
        "image": image_name,
        "width": width,
        "height": height,
        "conversations": conversations
    }
    entries_to_save.append(entry_to_save)
    idx +=1

with open('dset/prepared_data.jsonl', 'w') as f:
    for entry in entries_to_save:
        json_string = json.dumps(entry, separators=(',', ':'))
        f.write(json_string + '\n')
```
